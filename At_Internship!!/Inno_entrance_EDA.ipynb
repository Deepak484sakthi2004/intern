```python
import numpy as np
import pandas as pd
import seaborn as sns

df=pd.read_csv('Inno_entrance_test.csv')

print(df)

df.describe()
```

```python
df.corr()
```

There exists a +ve relation between input and output feature!

```python
df.info()
```

## UNI-VARIATE on INPUT feature!

```python
df['input'].plot(kind='kde')
```

```python
df['input'].plot(kind='box')
```

The input feature is clean and the mean is nearby 0

the input feature follows a gaussian distribution!

No outliers in the input feature

```python
# Statistical Analysis 

print('The mean of the input feature is         :',df['input'].mean())
print('The median of the input feature is       :',df['input'].median())
print('The mode of the input feature is         :',df['input'].mode()[0])

print('The average of the input feature is      :',df['input'].mean(),'-',df['input'].median())
print('The distribution of the input feature is :',df['input'].std())
```

There is a large difference between the mean and median, at the same time there seems to be no 
outliers in this feature!!

## UNI-VARIATE on OUTPUT feature!

```python
df['output'].plot(kind='kde')
```

```python
df['output'].plot(kind='box')
```

Presence of some outliers does not affect the data!!

```python
# Statistical Analysis 

print('The mean of the output feature is         :',df['output'].mean())
print('The median of the output feature is       :',df['output'].median())
print('The mode of the output feature is         :',df['output'].mode()[0])

print('The average of the output feature is      :',df['output'].mean(),'-',df['output'].median())
print('The distribution of the output feature is :',df['output'].std())
```
